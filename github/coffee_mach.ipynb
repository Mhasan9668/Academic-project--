```python
#calculater
import math
i='Addition'
ii='Substraction'
iii='Multiplication'
iv='Division'
v='factorial'
vi='power'
vii='sqrt'
operator=input('select a function that can you used')
# Addition (+)
if(operator == 'i'):
    first=float(input("enter first number"))
    second=float(input("enter second number"))
    print('the sum of', first ,'and', second,'is',first+second)
     # Substraction (-)
elif(operator == 'ii'):
    first=float(input("enter first number"))
    second=float(input("enter second number"))
    print('the difference of', first ,'and', second,'is',first-second)
     # Multiplication (*)
elif(operator == 'iii'):
    first=float(input("enter first number"))
    second=float(input("enter second number"))
    print('the multiple of', first ,'and', second,'is',first*second)
     # Division (/)
elif(operator == 'iv'):
    first=float(input("enter first number"))
    second=float(input("enter second number"))
    print('the division of', first ,'and', second,'is',first/second)
     # Modulo (%)
elif(operator == 'v'):
    first=int(input("enter first number"))
    a=math.factorial(first)
    print(a)
elif(operator == 'vi'):
    first=float(input("enter first number"))
    second=float(input("enter second number"))
    print('the power of', first ,'^', second,'is',first**second)
elif(operator == 'vii'):
    first=int(input("enter first number"))
    print('the sqrt of', first ,'is', math.sqrt(first))
     # Unknown
else:
     print ('Sorry, but I cannot understand your operation')
```
