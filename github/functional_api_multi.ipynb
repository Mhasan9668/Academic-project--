```python
from keras.layers import *
from keras.models import Model
```

```python
input1 = Input(shape=(32,))
input2 = Input(shape=(128,))
```

```python
hidden1 = Dense(32, activation="relu")(input1)
hidden2 = Dense(8, activation="relu")(hidden1)

output1 = Dense(1,activation='sigmoid')(hidden2)
```

```python
hidden3 = Dense(128, activation="relu")(input2)
hidden4 = Dense(64, activation="relu")(hidden3)
hidden5 = Dense(32, activation="relu")(hidden4)


output2 = Dense(1,activation='sigmoid')(hidden5)
```

```python
combined = concatenate([output1, output2])
```

```python
z = Dense(2, activation="relu")(combined)
z1 = Dense(1, activation="linear")(z)
```

```python
model = Model(inputs=[input1, input2], outputs=z1)
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```
